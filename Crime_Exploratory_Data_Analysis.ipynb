```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from wordcloud import WordCloud, STOPWORDS
%matplotlib inline
```

```python
df=pd.read_csv('Crime1.csv')
```

```python
df.head()
```

```python
df.Category.value_counts()
```

```python
category = pd.DataFrame(list(zip(df.Category.value_counts().index,df.Category.value_counts())), columns=['Category','value'], index=None)
```

```python
catcloud = WordCloud(
                          stopwords=STOPWORDS,
                          background_color='black',
                          width=1200,
                          height=800
                         ).generate(" ".join(category['Category'].values))
```

```python
plt.imshow(catcloud, alpha=0.8)
plt.axis('off')
plt.show()
```

```python
sb.catplot(x='value', y = 'Category', data=category,kind="bar", height=4.25, aspect=1.9, palette="cubehelix")
plt.title('Factorplot of the category of crime and number of occurences ')
```
